En el estudio de la programación y las matemáticas aplicadas, es fundamental aprender a traducir conceptos geométricos en algoritmos claros y funcionales. Este proyecto se centra en tres cálculos básicos pero esenciales: el área de un círculo, su perímetro (también conocido como circunferencia) y el volumen de una esfera. Las fórmulas matemáticas que sustentan estos cálculos son las siguientes:

1. 🔹 Área del círculo: A = π · r²
2. 🔹 Perímetro del círculo: P = 2 · π · r
3. 🔹 Volumen de la esfera: V = (4/3) · π · r³

![texto del vínculo](https://raw.githubusercontent.com/Eliseo128/Mi-Imagenes-de-EPPL-Unidad-2/refs/heads/main/circulo.PNG)



*   Donde r representa el radio de la figura y π (pi) es la constante matemática aproximadamente igual a 3.1415926535....
*   Para implementar estos cálculos de manera práctica, se ha desarrollado un programa en Python estructurado de forma modular.
* El programa utiliza un menú interactivo que se mantiene en ejecución mediante un ciclo while, permitiendo al usuario realizar múltiples cálculos sin reiniciar la aplicación.
*   La navegación se controla con estructuras condicionales if, elif y else, las cuales dirigen el flujo hacia la función correspondiente según la opción elegida.
*  Cada operación está encapsulada en su propia función (areacirculo(), perimetrocirculo() y volumenesfera()), cumpliendo con el principio de responsabilidad única y facilitando la lectura del código para estudiantes de nivel principiante.
*  Al invocar cada función, se solicita directamente el valor del radio, se aplica la fórmula matemática utilizando la librería estándar math y se muestra el resultado con un formato limpio y decorado con emojis
* Este enfoque no solo refuerza el pensamiento lógico y la sintaxis básica de Python, sino que también demuestra cómo la programación puede servir como herramienta práctica para resolver problemas geométricos de forma rápida y precisa.

```python
# ============================================================
# 📐 PROGRAMA DE CÁLCULOS GEOMÉTRICOS
# Elaborado por ✅ Eliseo Nava
# Fórmulas utilizadas:
#   Área del círculo    : A = π * r²
#   Perímetro del círculo: P = 2 * π * r
#   Volumen de la esfera : V = (4/3) * π * r³
# ============================================================

import math

def areacirculo():
    """Calcula y muestra el área de un círculo."""
    print("\n📐 Cálculo del Área del Círculo")
    radio = float(input("Ingresa el radio del círculo: "))
    area = math.pi * (radio ** 2)
    print(f"✅ El área del círculo con radio {radio} es: {area:.2f} unidades²\n")

def perimetrocirculo():
    """Calcula y muestra el perímetro (circunferencia) de un círculo."""
    print("\n📏 Cálculo del Perímetro del Círculo")
    radio = float(input("Ingresa el radio del círculo: "))
    perimetro = 2 * math.pi * radio
    print(f"✅ El perímetro del círculo con radio {radio} es: {perimetro:.2f} unidades\n")

def volumenesfera():
    """Calcula y muestra el volumen de una esfera."""
    print("\n🌐 Cálculo del Volumen de la Esfera")
    radio = float(input("Ingresa el radio de la esfera: "))
    volumen = (4/3) * math.pi * (radio ** 3)
    print(f"✅ El volumen de la esfera con radio {radio} es: {volumen:.2f} unidades³\n")

def menu():
    """Muestra el menú principal y gestiona la interacción del usuario."""
    while True:
        print("\n🧮 ═══════════════════════════════ 🧮")
        print("        📊 MENÚ DE CÁLCULOS 📊")
        print("🧮 ═══════════════════════════════ 🧮")
        print("1️⃣  Calcular área del círculo")
        print("2️⃣  Calcular perímetro del círculo")
        print("3️⃣  Calcular volumen de la esfera")
        print("4️⃣  Salir")
        print("🧮 ═══════════════════════════════ 🧮")

        opcion = input("👉 Selecciona una opción (1-4): ")

        if opcion == "1":
            areacirculo()
        elif opcion == "2":
            perimetrocirculo()
        elif opcion == "3":
            volumenesfera()
        elif opcion == "4":
            print("\n👋 ¡Gracias por usar el programa! Hasta pronto. 🚀\n")
            break
        else:
            print("\n⚠️ Opción no válida. Por favor, elige un número del 1 al 4.\n")

# Punto de entrada principal
if __name__ == "__main__":
    menu()
```
